## Problem 3

Let's assume the number of primes is finite i.e. the set of primes P = {$p_1,p_2,...,p_n$} has a finite cardinality (wlog, say n) and $p_i > 1$ $\forall$ $1 \leq i \leq n$. 

As per above assumption, there is a prime $p_n$, s.t. $p_n$ is the largest prime in the set P. 

Hence, $\forall x$, s.t. $x>p_n, x$ is composite.

Now, consider $x=(\Pi_{i=1}^{n} p_i) + 1$.
Clearly, $x > p_i$ $\forall$ $1 \leq i \leq n$ and hence, based on above assumption x is composite.

By Fundamental Theorem of Arithmetic, $\exists i$, s.t. $p_i | x$ and $p_i \in P$ i.e. $1 \leq i \leq n $, since x is composite.

$\implies p_i | (\Pi_{i=1}^{n} p_i) + 1$ 

But, $ p_i | (\Pi_{i=1}^{n} p_i)$

$\implies p_i | 1$ 

Also, $ 1|p_i$ as 1 is identity element. 

$\implies p_i = 1$ 

We arrive at a contradiction as we assumed, $p_i > 1$ $\forall$ $1 \leq i \leq n$.

Hence, $x=(\Pi_{i=1}^{n} p_i) + 1$ can't be composite, if the set P is finite.

So, $x=(\Pi_{i=1}^{n} p_i) + 1$ is prime, which is larger than $p_n$ and hence, $x=(\Pi_{i=1}^{n} p_i) + 1 \in P$ is the largest prime in that set P, which is a contradiction to the assumption that $p_n$ is the largest prime in finite set P.

Hence, the set P is not finite i.e. the number of primes is infinite.

$QED. $


## Problem 5

Consider the equation $x^2 = 5y - 2$.

Lets consider the same equation in modulo 5 group, $U_5$. Hence, taking modulo on both sides, the equation will reduce to:

$x^2 = 5y - 2$ mod $5. $

Since, $ 5y$ mod $5 = 0$ mod $5$ as $ 5 | 5y$.

$\implies x^2 = - 2$ mod $5. $

$ \implies x^2 = 3$ mod $5. $

Lets say, such an integral solution $x$ exists and satisfies above modularity equation.
Now, consider the congurent modulo 5 residue system. Any such number $x$ will have possible remainders in range {0,1,2,3,4} congurent modulo 5.

i.e., $x$ mod 5 $\in$ {$0,1,2,3,4$}.

Now, taking square, the possible values $x^2$ mod 5 can take will be in range {$0^2,1^2,2^2,3^2,4^2$}, i.e. {$0,1,4,9,16$} mod 5 = {$0,1,4,4,1$} mod 5.

i.e., $x^2$ mod 5 $\in$ {$0,1,4$}.

Which is a contraction to the assumption $x^2 = 3$ mod $5$ has such an integral solution $x$ as $3 \notin$ {$0,1,4$}.

Hence, the equation $x^2 = 5y - 2$ has no integral solution.

$QED.$
