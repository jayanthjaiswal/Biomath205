## Problem 5

Given,
\begin{equation}
A = \begin{pmatrix} 2 & -2\\ -2 & 5 \end{pmatrix}
\end{equation}

Since, A is a 2x2 square symmetric matrix and also, positive definite, we can find its unique Cholesky decomposition, i.e.

$A = LL^*$ , where L is a 2x2 lower triangular matrix with real and positive diagonal entries, and L* denotes the conjugate transpose of L.

Since, the entries in A are all real, conjugate transpose of L is equal to its transpose, i.e.

$L^* = L^T$

So then, above equation can be rewritten as,

$A = LL^T$

Now, assume a generic form of this 2x2 lower triangular matrix L, i.e.

\begin{equation}
L = \begin{pmatrix} a & 0\\ b & c \end{pmatrix}
\end{equation}

, where a, b, c are real and a>0, c>0 as diagonal entries are positive in L for Cholesky decomposition.

Plugging back it in the equation for Cholesky decomposition of A, we get

Given,
\begin{equation}
A = \begin{pmatrix} 2 & -2\\ -2 & 5 \end{pmatrix} = LL^T = \begin{pmatrix} a & 0\\ b & c \end{pmatrix}\begin{pmatrix} a & b\\ 0 & c \end{pmatrix}
\end{equation}

$
\implies \begin{pmatrix} 2 & -2\\ -2 & 5 \end{pmatrix} = \begin{pmatrix} a^2 & ab\\ ab & b^2+c^2 \end{pmatrix}
$

$
\implies a^2=2 , ab = -2, b^2+c^2 = 5.
$

$
\implies a=\sqrt 2 , b = - \sqrt 2, c = \sqrt 3
$
as L should have positive diagonal entries for Cholesky decomposition.

Hence, the Cholesky decomposition of A is,

\begin{equation}
A = \begin{pmatrix} 2 & -2\\ -2 & 5 \end{pmatrix} = LL^T = \begin{pmatrix} \sqrt 2 & 0\\ - \sqrt 2 & \sqrt 3 \end{pmatrix}\begin{pmatrix} \sqrt 2 & - \sqrt 2 \\ 0 & \sqrt 3 \end{pmatrix}
\end{equation}


## Problem 6

Given,
\begin{equation}
A = \begin{pmatrix} 1 & 2 & 2\\ 2 & 4 & 4\\ 2 & 4 & 17 \end{pmatrix}
\end{equation}

Since, A is a 3x3 square symmetric matrix and positive semi-definite, we can find its Cholesky decomposition $A = LL^*$ but there will be multiple such Cholesky decompositions as A is positive semidefinite (note that, $R_2 \rightarrow 2*R_1$).

Given,
\begin{equation}
B = \begin{pmatrix} 1 & 0 & 0\\ 2 & 0 & 0\\ 2 & 3 & 2 \end{pmatrix}
\end{equation}

B is a 3x3 square lower triangular matrix with positive diagonal elements.

Consider,
\begin{equation}
BB^T = \begin{pmatrix} 1 & 0 & 0\\ 2 & 0 & 0\\ 2 & 3 & 2 \end{pmatrix} \begin{pmatrix} 1 & 2 & 2\\ 0 & 0 & 3\\ 0 & 0 & 2 \end{pmatrix} = \begin{pmatrix} 1 & 2 & 2\\ 2 & 4 & 4\\ 2 & 4 & 17 \end{pmatrix} = A
\end{equation}

$
\implies BB^T = A.
$
Hence, B is a valid Cholesky decomposition of A.

Similarly given,
\begin{equation}
C = \begin{pmatrix} 1 & 0 & 0\\ 2 & 0 & 0\\ 2 & 0 & \sqrt 13 \end{pmatrix}
\end{equation}

C is also a 3x3 square lower triangular matrix with positive diagonal elements.

Consider,
\begin{equation}
CC^T = \begin{pmatrix} 1 & 0 & 0\\ 2 & 0 & 0\\ 2 & 0 & \sqrt 13 \end{pmatrix} \begin{pmatrix} 1 & 2 & 2\\ 0 & 0 & 0\\ 0 & 0 & \sqrt 13 \end{pmatrix} = \begin{pmatrix} 1 & 2 & 2\\ 2 & 4 & 4\\ 2 & 4 & 17 \end{pmatrix} = A
\end{equation}

$
\implies CC^T = A.
$
Hence, C is also a valid Cholesky decomposition of A.

There are actually multiple such Cholesky decompostion of A.
For example, take,
\begin{equation}
D = \begin{pmatrix} 1 & 0 & 0\\ 2 & 0 & 0\\ 2 & 1 & \sqrt 12 \end{pmatrix}
\end{equation}

D is also a 3x3 square lower triangular matrix with positive diagonal elements.

\begin{equation}
DD^T = \begin{pmatrix} 1 & 0 & 0\\ 2 & 0 & 0\\ 2 & 1 & \sqrt 12 \end{pmatrix} \begin{pmatrix} 1 & 2 & 2\\ 0 & 0 & 1\\ 0 & 0 & \sqrt 12 \end{pmatrix} = \begin{pmatrix} 1 & 2 & 2\\ 2 & 4 & 4\\ 2 & 4 & 17 \end{pmatrix} = A
\end{equation}

$
\implies DD^T = A.
$

So similarly, D is also a valid Cholesky decomposition of A.

Cholesky decomposition is not unique, when A is a positive semi-definite, which can easily be observed as its Gram matrix is not linearly independent (note that, $R_2 \rightarrow 2*R_1$ in A) or noticing that $Minor_{1,3} = Minor_{3,1} = 0$.
